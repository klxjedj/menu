```python
import json

d=json.load(open('./data.json','r',encoding='utf-8'))
```

```python
categoryData=[]
dishData=[]
for i in d:
    category={}
    category['categoryIndex']=i['categoryIndex']
    category['categoryName']=i['categoryName']
    category['itemsList']=[]
    for dish in i['items']:
        dishData.append(dish)
        category['itemsList'].append(dish['dishIndex'])
    categoryData.append(category)        
```
